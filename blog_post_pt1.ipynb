Why Meetup?
-------------

There are several things I like about Meetup as a dataset. In particular:

+ 100% of the data can be collected (as far as I can tell, I have access to everything that has ever happened on the site *except for deleted groups*
+ Everything is geolocated - people, events and groups all have associated locations
+ I don't have to go searching for events to know people interacted F2F, because they already exist
+ I don't have to go searching for groups and descriptions of them, because they already exist
+ People are able to represent themselves *differently to each group* with member profiles
+ We also have the data over time and in space.




```python
import pymongo, os, codecs, nltk, re, pprint
from datetime import datetime
from IPython.display import JSON, display_json, Image, HTML
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.spatial.distance import pdist
from __future__ import division
from pyquery import PyQuery as pq
from math import log, pow

%matplotlib inline

client = pymongo.MongoClient()
#databases:
dbs = ["buffalo"]
db_collections = ["events","group_member_profile", 
                  "groups","member_to_group","rsvps","venue", "members"]

db = client[dbs[0]]

```

```python
f = open("events.csv","w")
f.write("group_id,lat,lon,time,attendees,headcount\n")

for event in db.events.find():
    lat = -1
    lon = -1
    if event["venue_id"] is not None:
            venue = db.venue.find({"_id":event["venue_id"]},{"lat":1,"lon":1})[0]
            lat = str(venue['lat'])
            lon = str(venue['lon'])
    f.write(",".join([str(event["group_id"]),str(lat),str(lon),
                      str(datetime.fromtimestamp(event["time"]/1000)),
                      str(event['yes_rsvp_count']),
                      str(event['headcount'])])+"\n")
f.close()
```


What's the data?
-------------

The Meetup API gives you the ability to crawl as much data as you'd like, provided you play nicely with their more than reasonably rate limits.

Basically, the meetup data consists of four things I'm treating as first-class objects (or nodes, for network-y people): the users of the site, Meetup "groups", the events that are held by Meetup groups, and the "topics" that Meetup group creators specify their group with for searching purposes.  Members, groups and events all have additional properties that can be used to describe and understand them.  Topics are just words (e.g. "Python"), but are an interesting tool that can be used to explore the interconnections of the purposes of the different meetups.

From these first-class objects, we also get "networks" - connections between entities within these object groups.  For example, the "group to topic network" is a set of nodes and connections between those nodes (called "links") where a link is created whenever a group is labeled with a specific topic.  Other networks include the Member to Group network, the Group to Event network and the Member to Event network.  In fact, any combination of the four object groups creates a network, so there are twelve in total.  I'll only be focusing on one or two of them here.

```python
from collections import defaultdict
df = pd.DataFrame(list(db.groups.find({},{"lat":1,"lon":1, "topics":1})))

blah = []
counter = defaultdict(int)
for e in df.topics:
    blah.append([x["name"] for x in e])
    for x in e:
        
        counter[x["name"]] += 1

df["topic_list"] = ["|".join(x) for x in blah]

s = pd.Series(counter)
s.sort(ascending=False)
f = plt.figure(figsize=(10,8))
ax = f.add_subplot(1,1,1)
n = 50
ax.barh(range(n),s[:n].values)
ax.set_yticks(range(n))
z = ax.set_yticklabels(s.index[:n].values)
```

From personal experience and those of the friends I've spoken to, Meetup seems to be a hub for three far-from-mutually-exclusive things: making new friends, finding people to engage in hobbies that are off-the-beaten-path with ([drum circles](http://www.meetup.com/Southwestern-PA-Drum-Circle/) are pretty awesome, but that comes to mind here) and "networking", particularly with respect to the entrepeneur and technology communities.

This seems to hold somewhat true as we look at the histogram of the top 50 topics used to describe Meetup groups in Buffalo.  The four most popular terms all have to do with meeting new people and making friends, although admittedly these are also popular because these terms generalize to almost any situation. The idea of networking is well represented also, as terms like "Professional Development" and "Professional Networking" are scattered throughout the list.  

Finally, "off-the-beaten-path hobbies" is a bit of a catch-all, but things like spirituality and meditation could arguably fall under that category.  More realistically, however, it looks like the more popular terms relate not to off-the-beaten-path hobbies, but rather hobbies that are generally somewhat popular but may not be shared with individual's own social circle.  While Meetup claims to be a "neighborhood" centric model, this suggests the possibility that people are using Meetup to go outside their neighborhood, because they can't find anyone in their neighborhood who wants to meditate with them.  Naturally, without more analysis, this is just wild speculation.

```python
###Groups -> topics
g_to_t = codecs.open("groups_to_topics.csv", "w", encoding="utf-8")
g_to_t.write("Group,Topic\n")

for e in db.groups.find({},{"topics":1}):
    for t in e["topics"]:
        g_to_t.write(",".join([db.groups.find({"_id":e["_id"]},{"name":1})[0]["name"],t["name"]])+"\n")
g_to_t.close()

##Use ORA to generate the network image
Image(filename="images/gcc_lw2_hidependant_newman_selectivelabel.png")
```

While in truth, we only have data on the "Topic to Group network", we can capture a glimpse of the "Topic to Topic" network by multiplying the topic by group network by it's transpose. This gives us a glimpse into the way in which topics are shared between groups.  The image above, generated using ORA, shows the connections in the Topic to Topic network.  More specifically, a link between two topics is shown where those topics are both used to describe two or more Meetup groups.  

In the figure, nodes (and links) are colored by the different clusters of a run of the [Newman-Girvan algorithm](http://en.wikipedia.org/wiki/Girvan%E2%80%93Newman_algorithm). The resulting modularity value was high, ~.76, indicating that the clusters are relatively distinct in this figure.  Labels are only given for a select number of nodes, as showing all labels would just result in a big mess.

I only show the Largest Connected Component here, but we can see that within this faction of the network, clusters are not only distinct but relatively explainable.  In the bottom left, groups associated with software development are only linked to the rest of the topic network through the "Lean Startup" topic, which is in turn a term used to describe business and networking oriented groups, like "Social Networking".  There are two clusters of networking oriented groups (the light blue one and one of the green ones), one of which serves the Niagara region and one that is focused more on Buffalo.

Connected to the business and networking clusters through more general concepts of "Personal Growth" and "Health and Wellness" is a collection of "off-the-beaten-path" groups focusing on spirituality and health.  This makes sense in a stereotypical way - "business people" and people interested in "spirituality" do not normally mix, but both are interested in the more general topics of health and wellness.   Similarly, the two "business clusters" are connected to a cluster at the top of the image centered around outdoor activities via interstitial topics like "Adventure" and "Outdoor Fitness".  These common topics link the "outdoorsy people" to the "business people", but more "hardcore outdoorsy" things like "Nature Photography" are well separated from the business clusters.  Finally, a similar case can be made for the collection of topics on the left side of the figure, which all relate to Mothers and/or raising children, connected to the main business clusters via "Moms" and "Women's social".

From a social networks perspective, there's not much new here.  Homophily is one of, if not the, most established finding in the field.  Similarly, from a text-network perspective, common terms are expected to be both hubs and have high-betweenness, so common terms connecting different parts of the network is not entirely exciting.

The interesting part comes in, I think, when we compare the topic network to the "co-membership" network.  Intuitively, because the terms groups use to describe themeselves are clustered nicely, we would expect that the groups people go to are clustered as well.  In other words, I would have expected people who go to one "technology" group to, for the most part, only go to other technology groups.  


```python
g_to_m = codecs.open("groups_to_members.csv", "w", encoding="utf-8")
g_to_m.write("Group,Member\n")

for e in db.member_to_group.find():
    
    g_to_m.write(",".join([db.groups.find({"_id":e["group_id"]},{"name":1})[0]["name"],str(e["member_id"])])+"\n")
g_to_m.close()

###ORA

```

The figure above shows the co-membership network - that is, the network of Meetup groups in Buffalo, where a link is formed between two groups when X number of people are in both groups.  Here, there is a link only where there are more than 20 members in common, and I again only show the largest connected component.

Obviously, this analysis is horribly skewed by group size- in the next blog post, I'll be discussing how we can correct for this and get a more realistic picture of the degree to which two groups share the same members.  The big picture from the image above, however, is not likely to change. We can see that, in some cases, like the software dev groups at the bottom of the cluster, my intuition held true - these groups only share membership with other software companies.

For the most part, though, my intuition was horrendously wrong.  In reality, co-membership cannot be predicted by topic homophily alone because (some) software devs don't only like to write code, most mothers don't spend their entire lives worrying about their children, and darn it, my love of drum circles doesn't define me.  In short, the topic-topic network is easily predictable, but the co-membership network is not as easy, because people's interests are motivated by a much more complex set of factors.
